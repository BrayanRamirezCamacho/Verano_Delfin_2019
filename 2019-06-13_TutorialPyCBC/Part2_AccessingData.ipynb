## 2.1 Accessing LIGO/Virgo data

This will show the basics of how to get data from the LIGO/Virgo detectors

### Getting data around a  specific binary merger in the catalog

One can directly retrieve data around a specific GW even. Typically this data is centered on the event, though restrictions may apply which have not allowed this. This method by default gets the smallest version of the dataset. If additional data or specific versions are required, please see the further (possible following two additional) ways to access data.

```python
import pylab
from pycbc import catalog

GW150914 = catalog.Merger("GW150914")

# Get the time series data around GW150914 from the Hanford and Livingston detectors
H1       = GW150914.strain('H1')
L1       = GW150914.strain('L1')

# PILAS: H1 and L1 is a pycbc.types.TimeSeries object which contains
# gravitational-wave strain in this instance and has metadata
# such as the start time, and sample rate.

# We can see how much data was returned and its boundaries
# Note: All times are given in seconds since the GPS time epoch
print("Duration: {}s Start: {} End: {}".format(H1.duration, 
                                              int(H1.start_time),
                                              int(H1.end_time)))

# Plot the time series
pylab.plot(H1.sample_times,H1)
pylab.plot(L1.sample_times,L1)
pylab.ylabel('Strain')
pylab.xlabel('Time (s)')
pylab.show()
```
