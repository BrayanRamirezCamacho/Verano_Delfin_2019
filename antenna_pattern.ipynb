```python
from pycbc.detector import Detector
from pycbc.waveform import get_td_waveform
import numpy as np
import pylab

# Time, orientation and location of the source in the sky
pi = 3.14159
right_ascension = pi/4 
declination = pi/2
polarization = []
polarization = np.arange(0, 2*pi, 0.1)
inc = 0
time = 1000000000

# We can calcualate the antenna pattern for Hanford at
# the specific sky location
d = Detector("H1")
fp = []
fc = []
psi = []
# calculo de antenna pattern
# We get back the fp and fc antenna pattern weights.
for index in range(len(polarization)):
    f_p, f_c = d.antenna_pattern(right_ascension, declination, polarization[index], time)
    fp.append(f_p)
    fc.append(f_c)
    psi.append(polarization[index])
    print("Fp %g   Fc %g" % (fp[index], fc[index]))
pylab.plot(psi, fp, label = "Fp")
pylab.plot(psi, fc, label = "Fc")
pylab.legend()
pylab.grid()
pylab.xlabel("Psi")
pylab.ylabel("Fc, Fp")
pylab.show()
```

# Antenna Pattern T茅sis

```python
from pycbc.detector import Detector
from pycbc.waveform import get_td_waveform
import numpy as np
import pylab

#orientation and location of the source in the sky
pi = 3.14159
right_ascension = pi/4 
declination = pi/2
polarization = np.arange( 0, 2*pi, 0.1)
fp = []
fc = []
psi = []
for index in range(len(polarization)):
    fp.append(
                ((0.5) * (1 + pow(np.cos(declination),2)) * np.cos(2*right_ascension) * np.cos(2*polarization[index])) 
                - 
                (np.cos(declination) * np.sin(2*right_ascension) * np.sin(2*polarization[index]))
            )
    fc.append(
                ((0.5) * (1 + pow(np.cos(declination),2)) * np.cos(2*right_ascension) * np.sin(2*polarization[index])) + (np.cos(declination) * np.sin(2*right_ascension) * np.cos(2*polarization[index]))
            )
    psi.append(polarization[index])
    print("Fp %g Fc %g" % (fp[index], fc[index]))

pylab.plot(psi, fp, label="Fp")
pylab.plot(psi, fc, label="Fc")    
pylab.legend(loc='upper right')
pylab.legend()
pylab.grid()
pylab.xlabel("Psi")
pylab.ylabel("Fc, Fp")
pylab.show()
```

# Antenna Pattern para un detector en el Polo Norte

```python
from pycbc.detector import Detector
import pycbc
from pycbc.waveform import get_td_waveform
import numpy as np
import pylab

#orientation and location of the source in the sky
pi = 3.14159
right_ascension = pi/4 
declination = pi/2
polarization = np.arange( 0, 2*pi, 0.1)
fp = []
fc = []
psi = []
x = []
for index in range(len(polarization)):
    fp_aux, fc_aux = pycbc.detector.overhead_antenna_pattern(right_ascension, declination, polarization[index])
    fp.append(fp_aux)
    fc.append(fc_aux)
    print("Fp %g Fc %g" % (fp[index], fc[index]))
    x.append(polarization[index])

pylab.plot(x, fp, label = "Fp")
pylab.plot(x, fc, label = "Fc")
pylab.legend()
pylab.grid()
pylab.show()
```
