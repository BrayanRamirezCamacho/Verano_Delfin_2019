```python
# compute antenna pattern for LIGO and Virgo for highest incidence, lowest incidence,
# SN2017gax and SN2015as
```

```python
import numpy as np
import pycbc
from pycbc.detector import Detector
import GW_CCSNeFromCatalog
import GW_Template
import pylab

# choose a template
approx = 'And1' # ('And1','Dim1','Ott1','Yak1')

# read and plot raw template
# compute hp, hc from raw data
t, hp, hc, h = GW_CCSNeFromCatalog.Read(approx, 0)

# define scenarios
theta_2017, phi_2017, GPS_t_2017 = 4.17561569165, -3.40339204139, 1186765524# declination, ra, gps time
theta_2015, phi_2015, GPS_t_2015 = 2.23673211341, 2.61799387799, 1131580817#declination, ra, gps time
psi = 0

# declare detectors
h1 = Detector('H1')
h1_name = "Hanford"

# compute antenna pattern for every detector
h1_fp_2017, h1_fc_2017 = h1.antenna_pattern(phi_2017, theta_2017, psi, GPS_t_2017)
h1_fp_2017_2 = pow(h1_fp_2017, 2)
h1_fc_2017_2 = pow(h1_fc_2017, 2) 
h1_f_2017 = pow((h1_fp_2017_2 + h1_fc_2017_2), 0.5)

h1_fp_2015, h1_fc_2015 = h1.antenna_pattern(phi_2015, theta_2015, psi, GPS_t_2015)
h1_fp_2015_2 = pow(h1_fp_2015, 2)
h1_fc_2015_2 = pow(h1_fc_2015, 2) 
h1_f_2015 = pow((h1_fp_2015_2 + h1_fc_2015_2), 0.5)
    
print(h1_name,
      "SN2017gax",
      "Fp^2: %s, Fc^2: %s, F: %s" % (str(h1_fp_2017_2), str(h1_fc_2017_2), str(h1_f_2017)),
      "SN2015as",
      "Fp^2: %s, Fc^2: %s, F: %s" % (str(h1_fp_2015_2), str(h1_fc_2015_2), str(h1_f_2015)))

h1_h_2017 = hp * h1_fp_2017 + hc * h1_fc_2017
h1_h_2015 = hp * h1_fp_2015 + hc * h1_fc_2015

#print(h1_name,
#      "SN2017gax",
#      "h: %s " % (str(h1_h_2017)),
#      "SN2015as",
#      "h: %s " % (str(h1_h_2015)))

pylab.title("Hanford")
pylab.plot(t, h1_h_2017, label = "SN2017gax", color = 'g')
pylab.ylabel('h(t)')
pylab.xlabel('Time (s)')
pylab.grid()
pylab.legend()
pylab.show()

pylab.title("Hanford")
pylab.plot(t, h1_h_2015, label = "SN2015as", color = 'g')
pylab.ylabel('h(t)')
pylab.xlabel('Time (s)')
pylab.grid()
pylab.legend()
pylab.show()

# declare detectors
l1 = Detector('L1')
l1_name = "Livingston"

# compute antenna pattern for every detector
l1_fp_2017, l1_fc_2017 = l1.antenna_pattern(phi_2017, theta_2017, psi, GPS_t_2017)
l1_fp_2017_2 = pow(l1_fp_2017, 2)
l1_fc_2017_2 = pow(l1_fc_2017, 2) 
l1_f_2017 = pow((l1_fp_2017_2 + l1_fc_2017_2), 0.5)

l1_fp_2015, l1_fc_2015 = l1.antenna_pattern(phi_2015, theta_2015, psi, GPS_t_2015)
l1_fp_2015_2 = pow(l1_fp_2015, 2)
l1_fc_2015_2 = pow(l1_fc_2015, 2) 
l1_f_2015 = pow((l1_fp_2015_2 + l1_fc_2015_2), 0.5)
    
print(h1_name,
      "SN2017gax",
      "Fp^2: %s, Fc^2: %s, F: %s" % (str(l1_fp_2017_2), str(l1_fc_2017_2), str(l1_f_2017)),
      "SN2015as",
      "Fp^2: %s, Fc^2: %s, F: %s" % (str(l1_fp_2015_2), str(l1_fc_2015_2), str(l1_f_2015)))

l1_h_2017 = (-1)*(hp * l1_fp_2017 + hc * l1_fc_2017)
l1_h_2015 = (-1)*(hp * l1_fp_2015 + hc * l1_fc_2015)

#print(h1_name,
#      "SN2017gax",
#      "h: %s " % (str(h_2017)),
#      "SN2015as",
#      "h: %s " % (str(h_2015)))

pylab.title("Livingston")
pylab.plot(t, l1_h_2017, label = "SN2017gax", color = 'm')
pylab.ylabel('h(t)')
pylab.xlabel('Time (s)')
pylab.grid()
pylab.legend()
pylab.show()

pylab.title("Livingston")
pylab.plot(t, l1_h_2015, label = "SN2015as", color = 'm')
pylab.ylabel('h(t)')
pylab.xlabel('Time (s)')
pylab.grid()
pylab.legend()
pylab.show()

# declare detectors
v1 = Detector('V1')
v1_name = "Virgo"

# compute antenna pattern for every detector
v1_fp_2017, v1_fc_2017 = v1.antenna_pattern(phi_2017, theta_2017, psi, GPS_t_2017)
v1_fp_2017_2 = pow(v1_fp_2017, 2)
v1_fc_2017_2 = pow(v1_fc_2017, 2) 
v1_f_2017 = pow((v1_fp_2017_2 + v1_fc_2017_2), 0.5)

v1_fp_2015, v1_fc_2015 = v1.antenna_pattern(phi_2015, theta_2015, psi, GPS_t_2015)
v1_fp_2015_2 = pow(v1_fp_2015, 2)
v1_fc_2015_2 = pow(v1_fc_2015, 2) 
v1_f_2015 = pow((v1_fp_2015_2 + v1_fc_2015_2), 0.5)
    
print(h1_name,
      "SN2017gax",
      "Fp^2: %s, Fc^2: %s, F: %s" % (str(v1_fp_2017_2), str(v1_fc_2017_2), str(v1_f_2017)),
      "SN2015as",
      "Fp^2: %s, Fc^2: %s, F: %s" % (str(v1_fp_2015_2), str(v1_fc_2015_2), str(v1_f_2015)))

v1_h_2017 = (-1) * (hp * v1_fp_2017 + hc * v1_fc_2017)
v1_h_2015 = hp * v1_fp_2015 + hc * v1_fc_2015

#print(h1_name,
#      "SN2017gax",
#      "h: %s " % (str(h_2017)),
#      "SN2015as",
#      "h: %s " % (str(h_2015)))

pylab.title("Virgo")
pylab.plot(t, v1_h_2017, label = "SN2017gax", color = 'y')
pylab.ylabel('h(t)')
pylab.xlabel('Time (s)')
pylab.grid()
pylab.legend()
pylab.show()

pylab.title("Virgo")
pylab.plot(t, v1_h_2015, label = "SN2015as", color = 'y')
pylab.ylabel('h(t)')
pylab.xlabel('Time (s)')
pylab.grid()
pylab.legend()
pylab.show()

pylab.title("SN2017gax")
pylab.plot(t, h1_h_2017, label = "Hanford", color = 'g')
pylab.plot(t, l1_h_2017, label = "Livingston", color = 'm')
pylab.plot(t, v1_h_2017, label = "Virgo", color = 'y')
pylab.ylabel('h(t)')
pylab.xlabel('Time (s)')
pylab.grid()
pylab.legend()
pylab.show()

pylab.title("SN2015as")
pylab.plot(t, h1_h_2015, label = "Hanford", color = 'g')
pylab.plot(t, l1_h_2015, label = "Livingston", color = 'm')
pylab.plot(t, v1_h_2015, label = "Virgo", color = 'y')
pylab.ylabel('h(t)')
pylab.xlabel('Time (s)')
pylab.grid()
pylab.legend()
pylab.show()

# zoom
pylab.xlim(0.19, 0.22)
pylab.title("SN2017gax")
pylab.plot(t, h1_h_2017, label = "Hanford", color = 'g')
pylab.plot(t, l1_h_2017, label = "Livingston", color = 'm')
pylab.plot(t, v1_h_2017, label = "Virgo", color = 'y')
pylab.ylabel('h(t)')
pylab.xlabel('Time (s)')
pylab.grid()
pylab.legend()
pylab.show()

pylab.xlim(0.19, 0.22)
pylab.title("SN2015as")
pylab.plot(t, h1_h_2015, label = "Hanford", color = 'g')
pylab.plot(t, l1_h_2015, label = "Livingston", color = 'm')
pylab.plot(t, v1_h_2015, label = "Virgo", color = 'y')
pylab.ylabel('h(t)')
pylab.xlabel('Time (s)')
pylab.grid()
pylab.legend()
pylab.show()
```
